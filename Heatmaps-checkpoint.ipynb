```python
# This block generates the lat and lng by making gmaps api calls, currently set up to fill 'Lat' and 'Lng' 
params = {"key": gkey}

# Loop through rows and run a lat/long search for each city
for index, row in df_cs.iterrows():
    base_url = "https://maps.googleapis.com/maps/api/geocode/json"

    city = row.name[1]
    state = row.name[0]
    
    # update address key value
    params['address'] = f"{city},{state}"

    # make request
    cities_lat_lng = requests.get(base_url, params=params)
    
    print(city, state)
    #print(cities_lat_lng.url)
    
    # convert to json
    cities_lat_lng = cities_lat_lng.json()
    #pprint.pprint(cities_lat_lng)
    
    try:
        print(f'Latitude: {cities_lat_lng["results"][0]["geometry"]["location"]["lat"]}, Longitude: {cities_lat_lng["results"][0]["geometry"]["location"]["lng"]}')
        df_cs.loc[index, "Lat"] = cities_lat_lng["results"][0]["geometry"]["location"]["lat"]
        df_cs.loc[index, "Lng"] = cities_lat_lng["results"][0]["geometry"]["location"]["lng"]
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
        logging.exception()
        df_cs.loc[index, "Lat"] = np.NaN
        df_cs.loc[index, "Lng"] = np.NaN
    print("------------")         
# Visualize to confirm lat lng appear
df_cs.head()
```

```python
# Shows where cars are sold
fig = gmaps.figure()

heat_layer = gmaps.heatmap_layer(locations, weights=carprice.values, 
                                 dissipating=True, max_intensity=100,
                                 point_radius = 1)

# Adjust heat_layer setting to help with heatmap dissipating on zoom
heat_layer.dissipating = True
heat_layer.max_intensity = 40000
heat_layer.point_radius = 12

fig.add_layer(heat_layer)

fig
```
